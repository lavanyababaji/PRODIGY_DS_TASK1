```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/lavan/Downloads/all_weekly_excess_deaths.csv')
```

```python
df
```

```python
country_deaths = df.groupby('country')['total_deaths'].sum().sort_values(ascending=False)
```

```python
country_deaths
```

```python
plt.figure(figsize=(12, 8))
country_deaths.plot(kind='bar', color='orange')
plt.title('Total death by Countrty')
plt.xlabel('country')
plt.ylabel('Total Deaths')
plt.xticks(rotation=90)
plt.show
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing 'total_deaths' column

plt.figure(figsize=(12, 8))
plt.hist(df['total_deaths'].dropna(), bins=30, color='orange', edgecolor='green')  # Use dropna() to drop NaN values if any
plt.title('Distribution of Total Deaths')
plt.xlabel('Total Deaths')
plt.ylabel('Frequency')
plt.show()
```
